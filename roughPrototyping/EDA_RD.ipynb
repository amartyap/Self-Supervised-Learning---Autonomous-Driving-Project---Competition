```
# Analysis of Annotation.csv
import pandas as pd 
```

```
data = pd.read_csv("annotation.csv") 
```

```
data.head()
```

```
## check for imbalance: ( checking bounding boxes category) - overall

import collections

counts_bb = collections.Counter(data['category'])
new_list_count = sorted(counts_bb.items(), key = lambda x:x[1], reverse = True)
#new_lis
bbDF = pd.DataFrame(new_list_count, columns=['category', 'Frequency'])
bbDF
```

```
## check for imbalance: (checking action) - overall

counts_action = collections.Counter(data['action'])
newList_countAcc = sorted(counts_action.items(), key = lambda x:x[1], reverse = True)
#new_lis
nwCADF = pd.DataFrame(newList_countAcc, columns=['category', 'Frequency'])
nwCADF
```

```
## check for imbalance: ( checking bounding boxes category) -  per scene, per sample
dr = data.groupby(['scene', 'sample', 'category'])['category'].size()
dr.head(40)
dr.to_csv('sceneSampleCategory.csv', sep='\t', encoding='utf-8')
```

```
## check for imbalance: ( checking action) -  per scene, per sample
dr1 = data.groupby(['scene', 'sample', 'action'])['action'].size()
dr1.to_csv('sceneSampleAction.csv', sep='\t', encoding='utf-8')
dr1.head(40)
```
